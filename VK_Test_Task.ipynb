```python
import numpy as np
import pandas as pd
from catboost import Pool, FeaturesData, CatBoostRanker
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import ndcg_score

df = pd.read_csv('/kaggle/input/rec-data/intern_task.csv')
df_features = df.columns[2:].tolist()  # все фичи кроме rank и query_id
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

grouped_train = train_df.groupby('query_id')
train_pools = []
print('Forming train...')
for group in tqdm(grouped_train):
    group_df = group[1].sort_values('rank', ascending=False)
    features = group_df[df_features]
    labels = group_df['rank']
    pool = Pool(data=features, label=labels)
    train_pools.append(pool)

train_data = []
train_labels = []
print('Forming train pools...')
for pool in tqdm(train_pools):
    data = pool.get_features()
    label = pool.get_label()
    train_data.append(data[:])
    train_labels.append(label[:])

train_data = np.vstack(train_data)
train_labels = np.hstack(train_labels)
train_features = FeaturesData(train_data)
train_pool = Pool(train_features, train_labels)

model = CatBoostRanker(iterations=1000, random_seed=42, loss_function='RMSE', learning_rate=0.5)
model.fit(train_pool)

grouped_test = test_df.groupby('query_id')
test_pools = []
print('Forming test...')
for group in tqdm(grouped_test):
    group_df = group[1].sort_values('rank', ascending=False)
    features = group_df[df_features]
    test_pools.append(Pool(data=features))

print('Forming test pools...')
test_predictions = []
for pool in tqdm(test_pools):
    test_features = pool.get_features()
    test_features_data = FeaturesData(test_features)
    preds = model.predict(test_features_data)
    test_predictions.append(preds)

test_ranks = []
for preds in test_predictions:
    ranks = np.argsort(np.argsort(preds)) + 1
    test_ranks.append(ranks)

ndcg_scores = []
for ranks, group in zip(test_ranks, grouped_test):
    labels = group[1]['rank'].values

    mlb = MultiLabelBinarizer(classes=np.arange(1, ranks.max() + 1))
    labels_multilabel = mlb.fit_transform(labels.reshape(-1, 1))
    ranks_multilabel = mlb.transform(ranks.reshape(-1, 1))

    ndcg_scores.append(ndcg_score(labels_multilabel, ranks_multilabel, k=5))

print(f'Mean NDCG@5: {np.mean(ndcg_scores)}')
```

```python
import numpy as np
import pandas as pd
from catboost import Pool, CatBoostRanker
from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import ndcg_score

# Загрузка и подготовка данных
df = pd.read_csv('/kaggle/input/rec-data/intern_task.csv')
df_features = df.columns[2:].tolist()  # все фичи кроме rank и query_id

# Разделение на train и test
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Удаление групп с одним экземпляром
train_df = train_df.groupby('query_id').filter(lambda x: len(x) > 1)

# Подготовка данных для обучения
grouped_train = train_df.groupby('query_id')
train_pools = []
print('Forming train...')
for group in tqdm(grouped_train):
    group_df = group[1].sort_values('rank', ascending=False)
    features = group_df[df_features]
    labels = group_df['rank']
    pool = Pool(data=features, label=labels, group_id=[group[0]] * len(group_df))
    train_pools.append(pool)

# Объединение train pools
train_data = []
train_labels = []
print('Forming train pools...')
for pool in tqdm(train_pools):
    data = pool.get_features()
    label = pool.get_label()
    train_data.append(data[:])
    train_labels.append(label[:])

train_data = np.vstack(train_data)
train_labels = np.hstack(train_labels)
train_pool = Pool(train_data, train_labels)

# Обучение модели
model = CatBoostRanker(iterations=1000, random_seed=42, loss_function='YetiRank', learning_rate=0.5)
model.fit(train_pool)

# Подготовка данных для теста
grouped_test = test_df.groupby('query_id')
test_pools = []
print('Forming test...')
for group in tqdm(grouped_test):
    group_df = group[1].sort_values('rank', ascending=False)
    features = group_df[df_features]
    test_pools.append(Pool(data=features, group_id=[group[0]] * len(group_df)))

print('Forming test pools...')
# Получение предсказаний и присвоение рангов
test_predictions = []
for pool in tqdm(test_pools):
    test_features = pool.get_features()
    preds = model.predict(test_features)
    test_predictions.append(preds)

test_ranks = []
for preds in test_predictions:
    ranks = np.argsort(np.argsort(preds)) + 1
    test_ranks.append(ranks)

ndcg_scores = []
for ranks, group in zip(test_ranks, grouped_test):
    labels = group[1]['rank'].values
    ndcg = ndcg_score([labels], [ranks], k=5)
    ndcg_scores.append(ndcg)

print(f'Mean NDCG@5: {np.mean(ndcg_scores)}')

```
